## Self defined function to calculate positive, negative, and tone - HIV4 Dictionary

```python
hiv4_df = pd.read_csv('/Users/sheng/PycharmProjects/Training/Crowdfunding/IndiaGogo/HIV-4.csv', encoding='UTF-8')
```

```python
hiv4_positive = hiv4_df[['Entry', 'Positiv']].dropna()
hiv4_positive
```

```python
hiv4_negative = hiv4_df[['Entry', 'Negativ']].dropna()
hiv4_negative
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# procesing df.text data: tokenization, lemmatization, lower case, but keep stop words
def is_token_allowed(token):
    '''
        Only allow valid tokens which are not stop words
        and punctuation symbols.
    '''
    if (not token or not token.text.strip() or token.is_punct):
        return False
    return True

def preprocess_token(token):
    # Reduce token to its lowercase lemma form
    return token.lemma_.strip().lower()

# remove puctuation but keep stop words
processed_test = []

for i in df.text:
    complete_filtered_tokens = [preprocess_token(token) for token in nlp(i) if is_token_allowed(token)]
    processed_test.append(complete_filtered_tokens)
```

```python
len(processed_test)
```

```python
# add column
df['processed_text'] = processed_test
```

```python
# num_words_no_puctuation, but has stop words
num_words_no_puctuation = []

for i in df['processed_text']:
    num_words_no_puctuation.append(len(i))
```

```python
# add column
df['num_words_no_puctuation'] = num_words_no_puctuation
```

```python
df.columns
```

```python
pd.DataFrame([df['num_words_no_puctuation'], df['num_words'], df['num_words_spacy']]).T
```

```python
df.describe().T
```

```python
# make dictionary as one string
hiv4_pos_words = hiv4_positive['Entry']
hiv4_neg_words = hiv4_negative['Entry']
```

```python
hiv4_pos_words_list = ''
hiv4_neg_words_list = ''

for i in hiv4_pos_words:
    hiv4_pos_words_list += f'{ i} '
for j in hiv4_neg_words:
    hiv4_neg_words_list += f'{ j} '
```

```python
print(hiv4_pos_words_list)
```

```python
print(hiv4_neg_words_list)
```

```python
# process dictionary data with same froms as df.text 
processed_hiv4_pos_words = [preprocess_token(token) for token in nlp(hiv4_pos_words_list) if is_token_allowed(token)]
processed_hiv4_neg_words = [preprocess_token(token) for token in nlp(hiv4_neg_words_list) if is_token_allowed(token)]
```

```python
processed_hiv4_pos_words
```

```python
processed_hiv4_pos_words
```

```python
print(len(processed_hiv4_pos_words), len(processed_hiv4_neg_words))
```

```python
# only keep words, remove #1,2,3.....
hiv4_processed_pos_tokens = []
hiv4_processed_neg_tokens = []

for i in processed_hiv4_pos_words:
    if '#' in i:
        hiv4_processed_pos_tokens.append(i[:i.find('#')])
    else:
        hiv4_processed_pos_tokens.append(i)
        
for i in processed_hiv4_neg_words:
    if '#' in i:
        hiv4_processed_neg_tokens.append(i[:i.find('#')])
    else:
        hiv4_processed_neg_tokens.append(i)
```

```python
print(len(hiv4_processed_pos_tokens), len(hiv4_processed_neg_tokens))
```

```python
hiv4_processed_pos_tokens
```

```python
hiv4_processed_neg_tokens
```

```python
# find and count positive and negative words
num_pos = []
num_neg = []

for i in df.processed_text:
    pos = 0
    neg = 0
    
    for j in i:
        if j in hiv4_processed_pos_tokens:
            pos += 1
        if j in hiv4_processed_neg_tokens:
            neg += 1
    num_pos.append(pos)
    num_neg.append(neg)
```

```python
print(len(num_pos), len(num_neg))
```

```python
print(num_pos[:10])
print(num_neg[:10])
```

```python
# add new columns
df['hiv4_self_num_pos'] = num_pos
df['hiv4_self_num_neg'] = num_neg
```

```python
# df.drop(['self_num_pos', 'self_num_neg'], axis=1, inplace=True)
```

```python
df.columns
```

```python
# add new column: otne
df['self_hiv4_tone'] = (df['hiv4_self_num_pos'] - df['hiv4_self_num_neg']) / df['num_words_no_puctuation']
```

```python
# compare the package and self defined functions
pd.DataFrame([df.pysentiment2_hiv4_positive, df.hiv4_self_num_pos, df.pysentiment2_hiv4_negative, df.hiv4_self_num_neg]).T
```
